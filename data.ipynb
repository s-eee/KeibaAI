```python
import pandas as pd
import glob

# 元データファイルを結合
files = sorted(glob.glob("/content/drive/MyDrive/競馬(公開)/csvデータ/元データ/*.csv"))
csv = []
for file in files:
    print(file)
    print(pd.read_csv(file).shape)
    csv.append(pd.read_csv(file))


merge_list = pd.concat(csv, axis=0)
merge_list.to_csv("data.csv")
```

```python
# 学習に不要な行を削除
csv = pd.read_csv("data.csv")
csv = csv.drop("Unnamed: 0", axis=1)
csv = csv.drop("日時", axis=1)
csv = csv.drop("馬名", axis=1)
csv = csv.drop("馬番号", axis=1)
# ジョッキー名なしの場合
# csv = csv.drop("ジョッキー名", axis=1)

csv
```

```python
# 引退ジョッキーをダミー化から除外
# 現役
# "赤塚健（ばんえい）",
# "阿部武（ばんえい）",
# "金田利（ばんえい）",
# "菊池一（ばんえい）",
# "島津新（ばんえい）",
# "鈴木恵（ばんえい）",
# "竹ケ原（ばんえい）",
# "中山直（ばんえい）",
# "長澤幸（ばんえい）",
# "西謙一（ばんえい）",
# "西将太（ばんえい）",
# "林康文（ばんえい）",
# "藤野俊（ばんえい）",
# "藤本匠（ばんえい）",
# "船山蔵（ばんえい）",
# "松田道（ばんえい）",
# "松本秀（ばんえい）",
# "村上章（ばんえい）",
# "渡来心（ばんえい）",

# 減量騎手
csv = csv.replace('\n☆渡来心(ばんえい)\n', '\n渡来心(ばんえい)\n')
csv = csv.replace('\n☆竹ケ茉(ばんえい)\n', '\n竹ケ茉(ばんえい)\n')
csv = csv.replace('\n☆菊池一(ばんえい)\n', '\n菊池一(ばんえい)\n')
csv = csv.replace('\n☆西将太(ばんえい)\n', '\n西将太(ばんえい)\n')
csv = csv.replace('\n☆赤塚健(ばんえい)\n', '\n赤塚健(ばんえい)\n')
csv = csv.replace('\n☆金田利(ばんえい)\n', '\n金田利(ばんえい)\n')

# 引退
csv = csv.replace('\n入澤和(ばんえい)\n', '引退')
csv = csv.replace('\n☆井上真(ばんえい)\n', '引退')
csv = csv.replace('\n☆舘澤直(ばんえい)\n', '引退')
csv = csv.replace('\n☆貝羽智(ばんえい)\n', '引退')
csv = csv.replace('\n大口泰(ばんえい)\n', '引退')
csv = csv.replace('\n大河和(ばんえい)\n', '引退')
csv = csv.replace('\n安部憲(ばんえい)\n', '引退')
csv = csv.replace('\n尾瀬馨(ばんえい)\n', '引退')
csv = csv.replace('\n山本正(ばんえい)\n', '引退')
csv = csv.replace('\n工藤篤(ばんえい)\n', '引退')
csv = csv.replace('\n浅田達(ばんえい)\n' '引退')
csv = csv.replace('\n澁谷益(ばんえい)\n', '引退')
csv = csv.replace('\n舘澤直(ばんえい)\n', '引退')
csv = csv.replace('\n藤島隆(ばんえい)\n', '引退')
```

```python
# ダミー変数化
data = pd.get_dummies(csv, columns=["天候", "ジョッキー名", "牡0牝1"])
# ジョッキー名なしの場合
# data = pd.get_dummies(csv, columns=["天候", "牡0牝1"])

data
```
